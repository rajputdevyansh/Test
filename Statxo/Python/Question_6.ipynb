```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('data.csv')
```

```python
data.groupby('Name')['Sales'].sum().plot(kind='pie',autopct='%1.1f%%')
plt.title('Percentage Contribution of Sales for each Employee')
plt.ylabel('')
plt.show()
```
