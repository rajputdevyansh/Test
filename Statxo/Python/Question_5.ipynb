```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data=pd.read_csv('data.csv')
```

```python
data.groupby('Region')['Sales'].nlargest(5).plot(kind='bar',figsize=(10,5))
plt.title('Top 5 Sales for each Region')
plt.ylabel('Sales')
plt.show()
```
